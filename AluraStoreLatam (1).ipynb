### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
def facturacion(tienda_):
  facturacion_tienda = tienda_['Precio'].sum()
  return facturacion_tienda
```

```python
print(' tienda: ', facturacion(tienda),'\n','tienda2: ', facturacion(tienda2),'\n','tienda3: ', facturacion(tienda3),'\n','tienda4: ', facturacion(tienda4))
```

# 2. Ventas por categoría

```python
def ventasC(tienda_):
  ventasC_tienda = tienda_.groupby('Categoría del Producto')['Precio'].sum()
  return ventasC_tienda
```

```python
print(' tienda: ', ventasC(tienda),'\n','tienda2: ', ventasC(tienda2),'\n','tienda3: ', ventasC(tienda3),'\n','tienda4: ', ventasC(tienda4))
```

# 3. Calificación promedio de la tienda


```python
def calificacion(tienda_):
  calificacion_tienda = tienda_['Calificación'].mean()
  return calificacion_tienda
```

```python
print(' tienda: ', calificacion(tienda),'\n','tienda2: ', calificacion(tienda2),'\n','tienda3: ', calificacion(tienda3),'\n','tienda4: ', calificacion(tienda4))
```

# 4. Productos más y menos vendidos

```python
def productos(tienda_):
  productos_tienda = tienda_.groupby('Producto')['Precio'].sum().sort_values(ascending=False)
  return productos_tienda
```

```python
print(' tienda: ', productos(tienda),'\n','tienda2: ', productos(tienda2),'\n','tienda3: ', productos(tienda3),'\n','tienda4: ', productos(tienda4))
```

# 5. Envío promedio por tienda

```python
def envio(tienda_):
  envio_tienda = tienda_['Costo de envío'].mean()
  return envio_tienda
```

```python
print(' tienda: ', envio(tienda),'\n','tienda2: ', envio(tienda2),'\n','tienda3: ', envio(tienda3),'\n','tienda4: ', envio(tienda4))
```

# 6. GRAFICOS


```python
import matplotlib.pyplot as plt
import seaborn as sns

ingresos = [
    facturacion(tienda),
    facturacion(tienda2),
    facturacion(tienda3),
    facturacion(tienda4)
]
nombres = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

plt.figure(figsize=(8,5))
sns.barplot(x=nombres, y=ingresos, palette="coolwarm")
plt.title("Ingresos Totales por Tienda")
plt.ylabel("Ingresos")
plt.xlabel("Tienda")
plt.tight_layout()
plt.show()


```

```python
fig, axes = plt.subplots(2, 2, figsize=(12,10))

tiendas = [tienda, tienda2, tienda3, tienda4]
nombres = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

for i, ax in enumerate(axes.flat):
    ventas_cat = tiendas[i].groupby("Categoría del Producto")["Precio"].sum()
    ventas_cat.plot(kind='pie', autopct='%1.1f%%', startangle=90, ax=ax, cmap='Set3')
    ax.set_title(f'Distribución por Categoría - {nombres[i]}')
    ax.set_ylabel('')

plt.tight_layout()
plt.show()

```

```python
fig, axes = plt.subplots(2, 2, figsize=(12,10))

for i, ax in enumerate(axes.flat):
    tienda_actual = tiendas[i].copy()
    tienda_actual['Calificación'] = pd.to_numeric(tienda_actual['Calificación'], errors='coerce')
    calif_cat = tienda_actual.groupby("Categoría del Producto")["Calificación"].mean().sort_values()
    calif_cat.plot(kind='barh', ax=ax, color='mediumpurple')
    ax.set_title(f'Calificación Promedio - {nombres[i]}')
    ax.set_xlabel('Calificación')
    ax.set_ylabel('Categoría')

plt.tight_layout()
plt.show()

```
